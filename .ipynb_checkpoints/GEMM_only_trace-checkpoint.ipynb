````python
def filter_data_frame(df):
   
    df = df.astype(float)
    
    # consider only compute kernels
    #df = df.loc[ df['task_type'] == 0 ] 
    
    # remove first run
    df = df.loc[ df['taskpool_id'] > 8 ] 
    
    df.reset_index(drop=True, inplace=True)
    

    
    return df



````

````python
def filter_data_frame_class(df, class_id):
   
    df = df.astype(float)
    
    # consider only compute kernels
    #df = df.loc[ df['task_type'] == 0 ] 
    
    # remove first run
    df = df.loc[ df['taskpool_id'] > 8 ] 
    df = df.loc[ df['tcid'] == class_id ] 

    
    df.reset_index(drop=True, inplace=True)
    

    
    return df



````
