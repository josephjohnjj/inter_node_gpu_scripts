```python
def filter_data_frame(df):
   
    df = df.astype(float)
    
    # consider only compute kernels
    df = df.loc[ df['task_type'] == 0 ] 

    # remove the first run (warmup run)
    df = df.loc[ df['taskpool_id'] == 6 ] 

    #Change all timimng values fron ns to ms
    
    df.reset_index(drop=True, inplace=True)
    
    df['exec_time'] = 0
    df['exec_time'] = df['exec_time_end'] - df['exec_time_start']
    df['exec_time'] = df['exec_time'] / 1000000

    df['first_stage_in_time'] = 0
    df['first_stage_in_time'] = df['first_stage_in_time_end'] - df['first_stage_in_time_start']
    df['first_stage_in_time'] = df['first_stage_in_time'] / 1000000

    # only for tasks migrated after stage in 
    df['sec_stage_in_time'] = 0
    df.loc[df['mig_status'] == 2, 'sec_stage_in_time'] = df['sec_stage_in_time_end'] - df['sec_stage_in_time_start']
    df['sec_stage_in_time'] = df['sec_stage_in_time'] / 1000000

    df['stage_out_time'] = 0
    df['stage_out_time'] = df['stage_out_time_end'] - df['stage_out_time_start']
    df['stage_out_time'] = df['stage_out_time'] / 1000000

    df['epilogue_time'] = 0
    df['epilogue_time'] = df['complete_time_end']  - df['complete_time_start']
    df['epilogue_time'] = df['epilogue_time'] / 1000000 

    # only available for task not migrated 
    df['first_wt'] = 0
    df.loc[(df['mig_status'] == 0), 'first_wt'] = df['first_stage_in_time_start'] - df['first_queue_time']
    df['first_wt'] = df['first_wt'] / 1000000

    # only for migrated tasks 
    df['sec_wt'] = 0
    df.loc[(df['mig_status'] == 1), 'sec_wt'] = df['first_stage_in_time_start'] - df['second_queue_time']
    df.loc[(df['mig_status'] == 2), 'sec_wt'] = df['sec_stage_in_time_start'] - df['second_queue_time']
    df['sec_wt'] = df['sec_wt'] / 1000000
    
    df['turn_around_time'] = 0.0
    # non-migrated tasks
    df.loc[df['mig_status'] == 0, 'turn_around_time'] = df['exec_time_end'] - df['first_queue_time']
    # migrated tasks 
    df.loc[(df['mig_status'] == 1) | (df['mig_status'] == 2), 'turn_around_time'] = df['exec_time_end'] - df['second_queue_time']
    
    df['cost'] = 0.0
    df.loc[df['mig_status'] == 1, 'cost'] = df['first_stage_in_time_end'] - df['select_time']
    df.loc[df['mig_status'] == 2, 'cost'] = df['sec_stage_in_time_end'] - df['select_time']
    
    df['progress_time'] = 0.0
    df['progress_time'] = df['complete_time_end'] - df['first_queue_time'] -  df['first_stage_in_time'] - df['exec_time'] - df['stage_out_time'] -  df['epilogue_time'] 
    
    df.reset_index(drop=True, inplace=True)
    
    return df



```
